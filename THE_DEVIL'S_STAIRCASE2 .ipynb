## Cantor-Lebesgue Function (The Devil's Staircase)

(In what follows I have used __Measure, Integration & Real Analysis__ by _Sheldon Axler_ and __Measure And Integral: An Introduction to Real Analysis__ by _Richard L. Wheeden, Antoni Zygmund_ as the main resources.)


We use __Cantor set__ and __Cantor-Lebesgue function__ to stretch our intuition when dealing with Lebesgue measure. We assume the reader is already familiar with Cantor set and its properties. In what follows we briefly review some properties of the Cantor set.

__Lemma__. _<font color=blue>The Cantor set $C$ is the set of numbers in $[0, 1]$ that have a base $3$ representation consisting of only $0s$ and $2s$.</font>_  


__Properties__:
- The Cantor set is a closed subset of $\mathbb{R}$.

- The Cantor set has Lebesgue measure $0$.

- The Cantor set contains no interval with more than one element.



__Definition__. _<font color=blue>The</font> __<font color=red>Cantor-Lebesgue function</font>__ <font color=blue>$L: [0, 1] \rightarrow [0,1]$ is defined by converting base $3$ representations into base $2$ representations as follows</font>_:

- _<font color=blue>If $x \in C$, then $L(x)$ is computed by replacing all $2s$ in the unique representation of $x$ by $1s$.</font>_


- _<font color=blue>If $x \in [0,1] \backslash C$, then $L(x)$ is computed by truncating $x$ after the first $1$ in its $3$ base representation, and replacing all the remaining $2s$ by $1s$.</font>_


Alternatively, we could also define the _Cantor-Lebesgue function_ as the limit of a sequence of functions: let $D_k=[0,1] \backslash C_k$, where $C_k$ is the Kth step in constructing $C$ by removing the middle thirds. $D_k$ consists of $2^k-1$ intervals $I_j^k$ (ordered from left to right). Define $L_k$ to be a continuous function on $[0,1]$ which satisfies $L_k(0)=0$, $L_k(1)=1$, $L_k(x)=j 2^{-k}$ on $I_j^k$ for $j=1, ..., 2^k-1$, and which is linear on each interval of $C_k$.

Each $L_k$ is increasing, $L_{k+1}=L_k$ on $I_j^k$ for $j=1,..., 2^k-1$, and $\left|L_k-L_{k+1} \right| < {2^{-k}}$. Thus the sum $\sum\limits_{k=0}^n \, (L_k-L_{k+1})$ converges uniformly on $[0, 1]$, which in turn implies that $\{L_k\}$ converges uniformly on $[0, 1]$. Define $L=\lim\limits_{k \rightarrow \infty}\, L_k$; then $L(0)=0$, $L(1)=1$, $L$ is continuous and increasing, and $L$ is constant on every removed interval in constructing $C$. This $L$ is called _Cantor-Lebesgue_ function, also known as _The Devil's Staircase_. 

__Definition__. _<font color=blue>A function $f$ for which $f'$ is zero a.e. in $[a, b]$ is said to be</font> __<font color=red>singular</font>__ <font color=blue>on $[a, b]$.</font>_



The Cantor-Lebesgue function is an example of a non-canstant singular function on $[0,1]$.

__Theorem__. _<font color=blue>If $f$ is both absolutely continuous and singular on $[a, b]$, then it is constant on $[a, b]$.</font>_

__Proof__. _See Wheeden-Zygmund for a proof._


From this theorem we immediately conclude an important property of the function $L$:

__Corollary__. _<font color=blue>The Cantor-Lebesgue function $L$ is an example of a continuous function which is not absolutely continuous.</font>_

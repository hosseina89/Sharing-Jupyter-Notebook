```julia
using Plots
n= 200
th = Array(0:6*pi/100:12*pi) # theta from 0 to 2pi ( + a little extra)

anim = @animate for i in 1:n
plt=plot3d(10,xlim=(-1.5,1.5),ylim=(-1.5,1.5),zlim=(0, 25), c=:red, size=(700, 700, 700))
    

t = th[1:i]
    
x = cos.(t/3)
y = sin.(t/3)
z = t/3
    
plot3d!(plt, cos.(t/3),sin.(t/3),t/3, c=:blue, camera=(0, 90), legend=false)
    

scatter!([x[end]],[y[end]],[z[end]], c=:red, markerstrokecolor=:red)
    

end

gif(anim)
```

```julia
using Plots
n= 200
th = Array(0:pi/100:2*pi) # theta from 0 to 2pi ( + a little extra)

anim = @animate for i in 1:n
plt=plot3d(10,xlim=(-1.5,1.5),ylim=(-1.5,1.5),zlim=(0, 20), c=:red, size=(700, 700, 700))
    

t = th[1:i]
    
x = cos.(t)
y = sin.(t)
z = 5*(x.+y).+5
    
plot3d!(plt, cos.(t),sin.(t),5*(cos.(t).+sin.(t)).+5, c=:blue, camera=(90, 15), legend=false)
    

scatter!([x[end]],[y[end]],[z[end]], c=:red, markerstrokecolor=:red)
    

end

gif(anim)
```
